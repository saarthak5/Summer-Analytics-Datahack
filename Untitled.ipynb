```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sbn
```

```python
training_features = pd.read_csv('training_set_features.csv')
training_labels = pd.read_csv('training_set_labels.csv')
```

```python
training_features.head()
```

```python
print(training_features.shape)
print(training_features.isnull().sum(axis=0))
```

```python
training_features.drop(['employment_industry', 'employment_occupation', 'health_insurance'], axis=1, inplace=True)
```

```python
sbn.heatmap(training_features.isnull(), cbar=False, yticklabels=False)
```

```python
training_features.dropna(subset=['xyz_concern', 'xyz_knowledge', 'behavioral_antiviral_meds', 'behavioral_avoidance', 'behavioral_face_mask', 'behavioral_wash_hands', 'behavioral_large_gatherings', 'behavioral_outside_home', 'behavioral_touch_face', 'household_adults', 'household_children'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

def encoder(col):
    le = LabelEncoder()
    unq = training_features[col].unique()
    if unq[-1] is np.nan:
        unq = unq[:-1]
    le.fit(unq)
    training_features[col] = training_features[col].apply(lambda x : le.transform([x])[0] if type(x) == str else x)
```

```python
for col in ('age_group', 'education', 'race', 'sex', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa'):
    encoder(col)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imp = IterativeImputer()
training_features[training_features.columns] = imp.fit_transform(training_features[training_features.columns])
```

```python
sbn.heatmap(training_features.isnull(), cbar=False, yticklabels=False)
```

```python
print(training_labels.shape)
cnt = 0
for idx, x in training_features.iterrows():
    while idx != cnt:
        training_labels.drop(cnt, axis=0, inplace=True)
        cnt += 1
    cnt += 1
print(training_labels.shape)
```

```python
m, n = training_features.shape
X = np.concatenate([np.ones((m, 1)), training_features], axis=1)
Y1 = training_labels['xyz_vaccine'].to_numpy(copy=True)
Y2 = training_labels['seasonal_vaccine'].to_numpy(copy=True)
```

```python
def f(X, theta):
    # print(X.dot(theta))
    return 1 / (1 + np.e ** -X.dot(theta))

def regress(X, Y, theta, iter, _alpha, _lambda):
    for _ in range(iter):
        grad = np.zeros(theta.shape)
        for i in range(round(m/4)):
            grad = grad + (f(X[i], theta) - Y[i])*X[i]
        # theta = theta - (_alpha/m) * (grad + _lambda*theta)
        theta = theta - (_alpha/m)*grad
    return theta
```

```python
theta1 = np.ones(n+1)
theta2 = np.ones(n+1)
iter = 100
_alpha = 1e-20
_lambda = 0.00001
theta1 = regress(X, Y1, theta1, iter, _alpha, _lambda)
theta2 = regress(X, Y2, theta2, iter, _alpha, _lambda)
```
